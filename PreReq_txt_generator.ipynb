```python
file_path = 'course_data.csv'

# Load the CSV file into a pandas DataFrame
course_data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
course_data.head()
```

```python
pre_reqs = course_data["pr"]
names = course_data["no"]
classes = len(names)
```

```python
# Making matrix for pre-reqs
PR = []

max_length = max([len(ast.literal_eval(pre_reqs.loc[i])) for i in range(len(pre_reqs))])
print(max_length)

for i in range(classes):
    
    # initialize class_PR to all zeroes
    class_PR = np.zeros((max_length, classes))
    
    # convert to nested list of string
    p = ast.literal_eval(pre_reqs.loc[i])
    
    # initially class_PR will have one row for each pre-req
    # we iterate through all pre req constraints
    for j in range(len(p)):
        constraint = p[j]
        
        # add 1 to list for each constraint
        for p_class in constraint:
            try: # can't find some of the classes
                ind = names.tolist().index(p_class)
                class_PR[j][ind] = 1
            except:
                continue
    PR.append(np.transpose(class_PR))
    
# sanity check
print(sum(PR[0][1156]))
PR = np.array(PR)
print(PR.shape)
```

```python
# export numpy array

PR = np.array(PR)

# reshaping the array from 3D 
# matrice to 2D matrice. 

PR_reshaped = PR.reshape(PR.shape[0], -1) 

np.savetxt("pre_reqs.txt", PR_reshaped)
```
