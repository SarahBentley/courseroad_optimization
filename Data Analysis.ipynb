```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
courses = pd.read_csv('course_data.csv')
reqs = pd.read_csv('6-4_requirements.csv')
```

# All Classes

```python
# % of courses with no rating

count = (courses['ra']==0.0).sum()
no_ra = count/len(courses)

print('Percent of courses with no rating: ', no_ra*100)
```

```python
plt.hist(courses.loc[courses['ra']!=0]['ra'], bins=30, edgecolor='black')
plt.title('Histogram of MIT Course Ratings')
plt.xlabel('Rating')
plt.ylabel('Courses')
```

```python
sns.histplot(courses.loc[courses['ra']!=0]['ra'], bins=30,  kde=True)
plt.title('Histogram of MIT Course Ratings')
plt.xlabel('Rating')
plt.ylabel('Courses')
```

```python
# % of courses with no hours

count = (courses['h']==0.0).sum()
no_h = count/len(courses)

print('Percent of courses with no hours: ', no_h*100)
```

```python
plt.hist(courses.loc[courses['h']!=0]['h'], bins=30, edgecolor='black')
plt.title('Histogram of MIT Reported Hours of Courses')
plt.xlabel('Hours')
plt.ylabel('Courses')
```

# Course 6

```python
# % of course 6 classes with no rating

count = ((courses['co']== '6') & (courses['ra']==0.0)).sum()
no_ra = count/((courses['co']== '6').sum())

print('Percent of course 6 courses with no rating: ', no_ra*100)
```

```python
sns.histplot(courses.loc[courses['ra']!=0].loc[courses['co']=='6']['ra'], bins=20)
plt.title('Histogram of MIT Course 6 Class Ratings')
plt.xlabel('Rating')
plt.ylabel('Classes')
```

```python
# % of course 6 classes with no hours

count = ((courses['co']== '6') & (courses['h']==0.0)).sum()
no_ra = count/((courses['co']== '6').sum())

print('Percent of course 6 courses with no hours: ', no_ra*100)
```

```python
sns.histplot(courses.loc[courses['h']!=0].loc[courses['co']=='6']['h'], bins=20)
plt.title('Histogram of MIT Course 6 Class Hours')
plt.xlabel('Hours')
plt.ylabel('Classes')
```

# Course 18

```python
# % of course 6 and 18 classes with no ratings
count = ((courses['co']== '18') & (courses['ra']==0.0)).sum()
no_ra = count/((courses['co']== '18').sum())

print('Percent of course 18 courses with no rating: ', no_ra*100)
```

```python
sns.histplot(courses.loc[courses['ra']!=0].loc[courses['co']=='18']['ra'], bins=20)
plt.title('Histogram of MIT Course 18 Class Ratings')
plt.xlabel('Rating')
plt.ylabel('Classes')
```

```python
# % of course 6 and 18 classes with no hours
count = ((courses['co']== '18') & (courses['h']==0.0)).sum()
no_ra = count/((courses['co']== '18').sum())

print('Percent of course 18 courses with no hours: ', no_ra*100)
```

```python
sns.histplot(courses.loc[courses['h']!=0].loc[courses['co']=='18']['h'], bins=20)
plt.title('Histogram of MIT Course 18 Class Hours')
plt.xlabel('Hours')
plt.ylabel('Classes')
```

# Course 18+6

```python
# % of course 6 or 18 classes with no rating

count = (((courses['co']== '18')|(courses['co']== '6')) & (courses['ra']==0.0)).sum()
no_ra = count/(((courses['co']== '18')|(courses['co']== '6')).sum())

print('Percent of course 18 courses with no rating: ', no_ra*100)
```

```python
sns.histplot(courses.loc[courses['ra']!=0].loc[courses['co'].isin(['6', '18'])]['ra'], bins=20)
plt.title('Histogram of MIT Course 6 and Course 18 Class Ratings')
plt.xlabel('Rating')
plt.ylabel('Classes')
```

```python
# % of course 6 or 18 classes with no hours

count = (((courses['co']== '18')|(courses['co']== '6')) & (courses['h']==0.0)).sum()
no_ra = count/(((courses['co']== '18')|(courses['co']== '6')).sum())

print('Percent of course 18 courses with no hours: ', no_ra*100)
```

```python
sns.histplot(courses.loc[courses['h']!=0].loc[courses['co'].isin(['6', '18'])]['h'], bins=20)
plt.title('Histogram of MIT Course 6 and Course 18 Class Hours')
plt.xlabel('Hours')
plt.ylabel('Classes')
```

# Classes Fulfilling 6-4 Major Requirements

```python
# % of course 6-4 classes with no ratings

columns_to_check = reqs.columns.difference(['no'])
mask = (reqs[columns_to_check] == 1).any(axis=1)
courses_64 = reqs[mask]

courses_64_2 = courses[courses['no'].isin(courses_64['no'].to_list())]

count = (courses_64_2['ra']==0.0).sum()
no_ra = count/len(courses_64_2)

print('Percent of 6-4 courses with no rating: ', no_ra*100)
```

```python
sns.histplot(courses_64_2.loc[courses['ra']!=0]['ra'], bins=20)
plt.title('Histogram of MIT Class Ratings for Classes Fulfulling the 6-4 Major')
plt.xlabel('Rating')
plt.ylabel('Classes')
```

```python
# % of course 6-4 classes with no hours

columns_to_check = reqs.columns.difference(['no'])
mask = (reqs[columns_to_check] == 1).any(axis=1)
courses_64 = reqs[mask]

courses_64_2 = courses[courses['no'].isin(courses_64['no'].to_list())]

count = (courses_64_2['h']==0.0).sum()
no_ra = count/len(courses_64_2)

print('Percent of 6-4 courses with no rating: ', no_ra*100)
```

```python
sns.histplot(courses_64_2.loc[courses['h']!=0]['h'], bins=20)
plt.title('Histogram of MIT Class Hours for Classes Fulfulling the 6-4 Major')
plt.xlabel('Hours')
plt.ylabel('Classes')
```

# Comparison

```python

plt.figure(figsize=(15, 10))  
plt.boxplot([courses.loc[courses['ra']!=0]['ra'],
             courses.loc[courses['ra']!=0].loc[courses['co']=='6']['ra'], 
             courses.loc[courses['ra']!=0].loc[courses['co']=='18']['ra'],
             courses_64_2.loc[courses['ra']!=0]['ra']])

# Adding titles and labels
plt.title('Class Rating Distributions')
plt.ylabel('Ratings')
plt.xticks([1, 2, 3,4], ['All Classes', 'Course 6 Classes', 'Course 18 Classes', 'Classes Fulfulling 6-4 Requirements'])

# Showing the plot
plt.show()
```

```python
plt.figure(figsize=(15, 10))  
plt.boxplot([courses.loc[courses['h']!=0]['h'],
             courses.loc[courses['h']!=0].loc[courses['co']=='6']['h'], 
             courses.loc[courses['h']!=0].loc[courses['co']=='18']['h'],
             courses_64_2.loc[courses['h']!=0]['h']])

# Adding titles and labels
plt.title('Class Hours Distributions')
plt.ylabel('Hours')
plt.xticks([1, 2, 3,4], ['All Classes', 'Course 6 Classes', 'Course 18 Classes', 'Classes Fulfulling 6-4 Requirements'])

# Showing the plot
plt.show()
```
